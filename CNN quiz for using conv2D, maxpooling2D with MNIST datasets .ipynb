```python
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data ()  #array  
```

```python
print (X_train.shape)
print (X_test.shape)
```

```python
print (y_train.shape)
print (y_test.shape)
```

```python
# tensor    tensor   
import numpy as np
X_train = X_train.reshape (X_train.shape[0], 28, 28, 1).astype (np.float32) / 255   #0~255   255   0~1
X_test = X_test.reshape (X_test.shape[0], 28, 28, 1).astype (np.float32) / 255
```

```python
# Class label One hot encoding ( tensor       )


from keras.utils import np_utils

y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
# early stopping  call back  
#ModelCheckpoint   Keras       


model_path ="/content/drive/My Drive/Colab_myworks/models/{epoch:02d}-{val_loss:.4f}.hdf5"
check_pointer = ModelCheckpoint(filepath = model_path, monitor = 'val_loss'
                               , verbose = 1, save_best_only = True)
early_stopper = EarlyStopping (monitor = 'val_loss', patience = 10)
```

```python
history = model.fit (X_train, y_train, validation_data = (X_test, y_test),
                    epochs = 30, batch_size = 200, verbose = 0, callbacks = [early_stopper, check_pointer])
```

```python
model.evaluate (X_test, y_test)[1]
```
